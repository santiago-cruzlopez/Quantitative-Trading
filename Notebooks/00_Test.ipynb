# Compute Mandelbrot set

The Mandelbrot set is created by iterating a simple function across the complex plane. Points that cycle back to the same value are included in the set, while those that diverge are not. When visualized on a computer with various colors representing different divergence rates, the outside points can form beautiful images. The boundary of the Mandelbrot set is a fractal curve of infinite complexity, revealing intricate details and miniature replicas of the set upon magnification.

$$ Z_{n+1}=Z^{2}_{n}+C $$

Source:
- https://math.libretexts.org/Bookshelves/Analysis/Complex_Analysis_-_A_Visual_and_Interactive_Introduction_(Ponce_Campuzano)/05%3A_Chapter_5/5.05%3A_The_Mandelbrot_Set

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(x, y, max_iter, escape=True):
    # Create a 2D array to hold the output
    output = np.zeros((len(y), len(x)))

    for i in range(len(y)):
        for j in range(len(x)):
            c = complex(x[j], y[i])
            z = 0
            n = 0
            
            while abs(z) <= 2 and n < max_iter:
                z = z*z + c
                n += 1
            
            # If escape is True, we store the number of iterations
            if escape:
                output[i, j] = n
            else:
                # If escape is False, we can store a boolean value
                output[i, j] = n < max_iter

    return output

# Define the range for x and y
x = np.linspace(-2.25, 0.75, 500)
y = np.linspace(-1.25, 1.25, 500)

# Generate the Mandelbrot set output
output = mandelbrot(x,y,200,False)


# Plot the output
plt.imshow(output, extent=(x.min(), x.max(), y.min(), y.max()), cmap='hot')
plt.colorbar()
plt.title("Mandelbrot Set")
plt.xlabel("Re")
plt.ylabel("Im")
plt.show()
```
