Below is the formula of the mathematical model of a neuron.

$$OUT = f(W,X) =f(\sum_{i=1}^{n}W_iX_i)$$

Where:
*   f = activation function
*   n = number of elements in the input sequence
*   W_i = weight of the ith element of the sequence
*   X_i = ith element of the input sequence

```python
#Take the average of the last 60 days at each timepoint
MAVG = pd.Series(X).rolling(60).mean()
plt.plot(X.index, X.values)
plt.plot(MAVG.index, MAVG.values, label='60 Day MA')
plt.ylabel('Close Price')
plt.legend(['INTC','60 Day MA']);
```

```python
#Example of three Moving Averages

MAVG1 = pd.Series(X).rolling(10).mean()
MAVG2 = pd.Series(X).rolling(30).mean()
MAVG3 = pd.Series(X).rolling(60).mean()
plt.plot(X.index, X.values)
plt.plot(MAVG1.index, MAVG1.values)
plt.plot(MAVG2.index, MAVG2.values)
plt.plot(MAVG3.index, MAVG3.values)
plt.ylabel('Close Price')
plt.legend(['INTC','10 Day MA','30 Day MA','60 Day MA']);
```

```python
# Set the start and end date
start_date = '1990-01-01'
end_date = '2025-01-23'

# Define the ticker list
tickers_list = ['AAPL', 'IBM', 'MSFT', 'WMT']

# Create placeholder for data
data = pd.DataFrame(columns=tickers_list)

# Fetch the data
for ticker in tickers_list:
    data[ticker] = yf.download(ticker,
                               start_date,
                               end_date)['Close']

# Print first 5 rows of the data
data.head()
```

```python
# Plot all the close prices
data.plot(figsize=(10, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Close Price", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
